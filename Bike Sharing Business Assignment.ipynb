### Step - 3: Data prepration

```python
Bikesharing.drop(['instant', 'dteday', 'weekday', 'temp', 'casual', 'registered'],axis=1,inplace=True)
```

```python
Bikesharing.head()
```

```python
month1 = pd.get_dummies(Bikesharing['mnth'],drop_first = True)
weathersit1 = pd.get_dummies(Bikesharing['weathersit'],drop_first = True)
season1 = pd.get_dummies(Bikesharing['season'],drop_first = True)
```

```python
Bikesharing = pd.concat([Bikesharing, month1,weathersit1,season1], axis = 1)
Bikesharing.drop(['mnth','weathersit','season'], axis = 1, inplace = True)
```

```python
Bikesharing.head()
```

### Spliting and Rescaling the Data

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(Bikesharing, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
listcat1 = ['season', 'mnth','weathersit']
listnum1 = ['yr', 'holiday', 'workingday', 'atemp', 'hum', 'windspeed', 'cnt']

df_train[listnum1] = scaler.fit_transform(df_train[listnum1])
```

```python
df_train.describe()
```

```python
plt.figure(figsize = (20, 12))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

### Build the Linear Model

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train)

LR_M1 = sm.OLS(y_train, X_train_lm).fit()

LR_M1.params
```

```python
LR_M1.summary()
```

```python
# Checking VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
